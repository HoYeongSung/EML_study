```python
# PI
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

#    ( )
def f(x): return np.sin(x) * 0.9 + 0.1 * x

#     ()
x = np.linspace(0, 10, 1000)
mu = f(x)              #  
sigma = 0.2 * np.ones_like(x)  #  

#   
f_best = 0.7
xi = 0.01

# PI 
z = (mu - f_best - xi) / sigma
pi = norm.cdf(z)

# 
plt.figure(figsize=(10, 5))
plt.plot(x, mu, label='GP mean ()', color='blue')
plt.fill_between(x, mu - sigma, mu + sigma, alpha=0.2, label='GP uncertainty ()')
plt.axhline(f_best + xi, color='gray', linestyle='--', label='Target threshold')

plt.plot(x, pi, label='PI(x)', color='red')
plt.xlabel('x')
plt.ylabel('PI and f(x)')
plt.legend()
plt.title('Probability of Improvement (PI)')
plt.show()
```
